```문자열 하나를 입력받아 인터넷 주소를 반환하는 make_url 함수를 만들어주세요```

출력 : www.google.com

<span style="color:red">답변</span>

def 함수1(num) :<br>
&emsp;&emsp;return 함수0(num + 2)<br>

def 함수2(num) :<br>

c = 함수2(2)<br>

````python
def 함수0(num) :
    return num * 2

def 함수1(num) :
    return 함수0(num + 2)

def 함수2(num) :
    num = num + 10
    return 함수1(num)

c = 함수2(2)
print(c)
````
